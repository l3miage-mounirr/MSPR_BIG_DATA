```python
# On importe les différents dataset des éléctions présidentielles
import pandas as pd
import numpy as np
import xlrd

# https://www.data.gouv.fr/fr/pages/donnees-des-elections/

df_2002 = pd.read_excel("C:/Users/Lassy/Cours/MSPR3/Présidentielle/Presidentielle_2002_Resultats_Tour_1.xls",sheet_name=0)
df_2007 = pd.read_excel("C:/Users/Lassy/Cours/MSPR3/Présidentielle/Présidentielle 2007 t1 et t2.xls",sheet_name=0)
df_2012 = pd.read_excel("C:/Users/Lassy/Cours/MSPR3/Présidentielle/Présidentielle 2012 t1 et t2.xls",sheet_name=0)
df_2017 = pd.read_excel("C:/Users/Lassy/Cours/MSPR3/Présidentielle/Presidentielle_2017_Resultats_Communes_Tour_1_c.xls",sheet_name=0,skiprows=3)
df_2022 = pd.read_excel("C:/Users/Lassy/Cours/MSPR3/Présidentielle/Presidentielle_2022_Resultats_Tour_1.xlsx")
```

```python
# Fonction pour le pivot des candidats en ligne
def pivot_and_concat(colonnes_pivot, colonnes_initiales, df_entree, annee):
    # Créez une liste vide pour stocker les DataFrames individuels
    dfs = []

    # Parcourez les colonnes pivot et créez un DataFrame pour chaque groupe de colonnes (Nom, Prénom, Voix)
    for i in range(0, len(colonnes_pivot), 3):
        groupe_colonnes = colonnes_pivot[i:i + 3]
        df_temp = df_entree[colonnes_initiales + groupe_colonnes].copy()
        df_temp.columns = colonnes_initiales + ['Nom', 'Prénom', 'Voix']
        dfs.append(df_temp)

    # Concaténez les DataFrames individuels en un seul
    df_filtre = pd.concat(dfs, ignore_index=True)

    # Réinitialisez l'index
    df_filtre.reset_index(drop=True, inplace=True)

    # On conserve uniquement le top 10 des villes de France sans Paris
    df_filtre=df_filtre.loc[df_filtre['Libellé de la commune'].isin(['Marseille','Lyon','Toulouse','Nice','Nantes','Montpellier','Strasbourg','Bordeaux','Lille','Rennes'])]


    # On fait la somme du nombre de voix d'un candidat pour la même ville car on avait le détail au niveau arrondissement
    df_somme = df_filtre.groupby(['Libellé de la commune','Nom','Prénom']).agg({'Voix': 'sum'}).reset_index()

    # Réinitialisez l'index du DataFrame final
    df_somme.reset_index(drop=True, inplace=True)


    # On va ajouter une colonne pour avoir l'année ce qui permettra de créer le dataset final
    df_somme = df_somme.assign(Année=annee)
    return df_somme
```
